## This notebook calculates the metrics for parent curation id = 1

```python
from spyglass.spikesorting.v0.spikesorting_curation import WaveformParameters,WaveformSelection,Waveforms,QualityMetrics,MetricSelection

from spyglass.spikesorting.v0 import (Curation)
from spyglass.utils.nwb_helper_fn import get_nwb_copy_filename
from spyglass.shijiegu.decodeHelpers import runSessionNames, sleepSessionNames
from spyglass.shijiegu.singleUnit import qualityMetric_master
```

```python
nwb_file_name = 'lewis20240107.nwb'
nwb_copy_file_name = get_nwb_copy_filename(nwb_file_name)

intervals, _ = runSessionNames(nwb_copy_file_name)
intervals2, _ = sleepSessionNames(nwb_copy_file_name)
print(intervals)
```

```python
for session_name in intervals:
    qualityMetric_master(nwb_copy_file_name,session_name,parent_curation_id = 1)
```

```python
for session_name in intervals:
    curation_num = len(Curation & {"nwb_file_name": nwb_copy_file_name,
                         "sorter":"mountainsort4",
                         "sort_interval_name":session_name,
                          "curation_id":1})
    quality_metric_num = len(QualityMetrics & 
                             {"nwb_file_name": nwb_copy_file_name,
                             "sorter":"mountainsort4",
                             "sort_interval_name":session_name,
                              "curation_id":1})
    assert curation_num == quality_metric_num
```
